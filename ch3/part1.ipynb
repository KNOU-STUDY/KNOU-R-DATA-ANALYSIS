```r
dat1 <- read_excel("../data/auto-mpg.xlsx")
```

```r
t(dat1)
```

```r
dat.base5<-dat1
```

```r
t(dat.base5)
```

```r
m <- mean(dat.base5$weight)
m
```

```r
# 두개로 범주를 나눌 때는 ifelse를 사용하면된다.
dat.base5$weight_rank <- ifelse(dat.base5$weight >= m, 1 , 0)
```

```r
t(dat.base5$weight_rank)
```

```r
# 몸무게를 5분위수로 나눠서 구하고 양끝에 최소(0)와 최대(Inf)를 넣는다.
q <- c(0)
for(i in seq(from=0, to=1, length=5)){
    q <- c(q, as.integer(quantile(dat.base5$weight, i)))
}
q <- c(q, Inf)
q
```

```r
dat.base5$weight_cut <- cut(dat.base5$weight, breaks=q)
t(as.integer(dat.base5$weight_cut))
```

```r
t(dat.base5$weight_cut)
```

```r
library(dplyr)
dat.dplyr5 <- dat.base5 %>% mutate(weight_rank=ifelse(weight>=m,2,0))
t(dat.dplyr5$weight_rank)
```

```r
library(dplyr)
dat.dplyr6 <- dat.base5 %>% mutate(weight_cut=cut(dat.base5$weight, breaks=q))
t(dat.dplyr6$weight_cut)
```

```r
t(as.integer(dat.dplyr6$weight_cut))
```

```r
summary(dat.dplyr6)
```

```r
t(dat.dplyr6)
```

```r
sz <- c(as.integer(count(dat.dplyr6)))
rk <- c('a','b','c','d','e','f')
dat.dplyr6$weight_alpha_rank <- 'NULL'
for (i in 1:sz){
    dat.dplyr6$weight_alpha_rank[i] <- rk[as.integer(dat.dplyr6$weight_cut)[i]+1]
}
t(dat.dplyr6$weight_alpha_rank)
```

```r
summary(dat.dplyr6)
```

```r
# 범주형자료로 변경
dat.dplyr6$weight_alpha_rank <- as.factor(dat.dplyr6$weight_alpha_rank)
```

```r
summary(dat.dplyr6)
```
