```r
# 예제 2.1
# A, B, C, D가 관찰될 확률이 각각 0,1, 0.2, 0.3 및 0.4인 모집단에서 10개를 복원추출로 뽑는 모의실험을 해 보자.
```

```r
x <- c('A', 'B', 'C', 'D')
p <- c(.1, .2, .3, .4)
sample(x, size=10, replace=T, prob=p)
```

```r
# 예제 2.2
# 1부터 1,000 사이의 자연수에게 비복원으로 10개의 난수를 추출하자.
```

```r
sample.int(1000, size=10)
```

```r
# 예제 2.3
# 공평한 동전을 1,000번 던져서 나ㅓ오는 앞면의 수를 위의 sample 또는 sample.int 함수를 사용하여 얻어 보자.
```

```r
# prob는 각각의 원소가 추출될 확률임, 즉 반반, replace는 복원추출
ntoss = sample(c(0,1), 1000, replace=T, prob=c(0.5, 0.5))
ntoss
```

```r
# sample.int에서 2보다 같거나 작은 자연수를 출력하는데 1을 빼니 0과 1이 된다.
ntoss2 = sample.int(2, 1000, replace=T) - 1
ntoss2
```

```r
sum(ntoss)
sum(ntoss2)
```

```r
# 예제 2.4
# 로또는 1부터 45 사이의 숫자를 순서에 상관없이 6개를 맞추면 1등이다.
# 5등은 3개를 맞추는 경우라고 한다. 만일 당첨번호가 1,2,3,4,5,6이라고 할 때 백만 번 모의실험을 하여 로또 5등에 당첨될 확률을 얻어보자.
```

```r
lotto <- function(nn=1000){
    luckyNo <- c(1, 2, 3, 4, 5, 6)
    threeNo <- 0
    for (i in 1:nn){
        x <- sort(sample.int(45, size=6))
        nMatch = 0
        for (j in 1:6){
            for(k in j:6){
                if(x[j]==luckyNo[k]){
                    nMatch = nMatch + 1
                }
            }
        }
        if(nMatch == 3) {
            threeNo = threeNo + 1
        }
    }
    list(threeNo=threeNo)
}
```

```r
lotto(10000)
```

```r
lotto2 <- function(nn=1000){
    luckyNo <- c(1, 2, 3, 4, 5, 6)
    threeNo <- 0
    for (i in 1:nn){
        x <- sort(sample.int(45, size=6))
        if(sum(luckyNo %in% x) == 3) {
            threeNo = threeNo + 1
        }
    }
    list(threeNo=threeNo)
}
```

```r
lotto2(10000)
```
